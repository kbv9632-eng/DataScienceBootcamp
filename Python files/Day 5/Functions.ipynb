# Functions


```python
# Functions - is a reusable block of code that executes a specific task only when it is called. A parameter is a variable defined in the function definition that acts as a placeholder for incoming values.It doesn't have a value yet. It's simply waiting for one.
#An argument is the actual value you pass into the function when you call it.
# Example 
def greet(name):
    print("Hello {0} ".format(name))

first_call = greet("Kiran")
```

```python
# Return : ketword used inside functions, that returns something, i.e  Python replacing the function call with the returned value.A return statement immediately ends the function.
# Example 
def add(a, b):
    print(a + b) # Only role of this inbuilt-func is to display something where as return does the opposite it gives but does not display . Basically the var that call the func call will have the value 

answer = add(10, 20)

print(answer)
```

# Postional Arguments 
positional arguments are values passed to a function in a specific sequence.

```python
# Parameters can have default values and those values can change if you pass the new value while calling teh function.But req parameters come before default parameters."A default parameter is used only when no argument is provided for that parameter during the function call. If an argument is provided, it overrides the default value."
def greet(name =' Guest'):
    print('Hello ',name)


greet()
greet('kiran')

def personal_info(name,age=18):
    print('Hello {0} and your age is {1} years old '.format(name,age))

personal_info('Kiran')
personal_info('Kiran',24)  
```

# Key word  Arguments 
You can also pass arguments by name i.e python lets us match by parameter name.

```python
# Example 
def introduce(name, age):
    print(name, age)

introduce(age=25, name="John") # Usefule when you have many parameters 
```

# Imp Rule 

You can mix positional and keyword arguments only if:  All positional arguments come first, followed by keyword arguments.
3 ways to call a func : - All Postional Args, All Keywords, Postional first and then keyword args.


```python
# Example
def student(name, age, city):
    print(name, age, city)

student(city="Bangalore", name="Kiran", age=24)
```
