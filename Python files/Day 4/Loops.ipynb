```python
# Here we are going to learn about Loops 
# Used to iterate through a data structure that contains info/data.
# Sample Loop 
players = ["Messi", "Ronaldo", "Neymar"]

for player in players:
    print(player)

print("    ")

for i in range(0,5):
    print(i)
```

```python
# Enumerate - whenever you are iterating through a list , it gives you teh index and and the value 
numbers = [15,42,8]

for index, num in enumerate(numbers):
    print(index, num)

```

```python
#Break - stops the loop immediately 
# for eg 
for i in range(3):
    print(i)
    break
```

```python
# Continue - Skip this iteration 
for i in range(5):

    if i==2:
        continue

    print(i)
```

```python
# pass - basically means ignore or in other words do nothing. Useful while still building your program 
for i in range(5):
    if i == 2:
        pass  # As you does nothing or else it throws an incomplete input error 


# Another example 

if True :
    pass
```

```python
# While Loop - Unlike a for loop, a while loop keeps running until the condition becomes False.
# for example

count = 1

while count <= 5:  # If forget the condition , you will stuck in an ifinite loop 
    print(count)
    count += 1
```

```python
# Nested Loops - basically a loop inside a loop.The inner loop finishes completely before the outer loop moves to the next iteration.
# for example 

for i in range(3):

    for j in range(2):

        print(i, j)

#Another example

for i in range(1,4):

    for j in range(1,4):

        print(i*j)
```
