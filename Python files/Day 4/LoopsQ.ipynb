```python
# Q3 
matrix = [
    [4, 8, 2],
    [7, 1, 5],
    [9, 6, 3]
]

largest_num_matrix = matrix[0][0]

for row in matrix :
    for number in row:
        if number > largest_num_matrix:
            largest_num_matrix = number 


print(largest_num_matrix)
```

```python
# Q4 
matrix = [
    [4, 8, 2],
    [7, 1, 5],
    [9, 6, 3]
]

even_num_count = 0

for row in matrix:
    for number in row:
        if number %2 == 0:
            even_num_count +=1

print(even_num_count)
```

```python
# Q10 

students = [
    {
        "name": "John",
        "marks": [85, 90, 78]
    },
    {
        "name": "Sarah",
        "marks": [92, 88, 95]
    },
    {
        "name": "Mike",
        "marks": [70, 65, 80]
    }
]

highest_avg_mark = 0
star_stud = " "


for student in students:
    nos_marks = 0
    total_marks = 0

    for marks in student['marks']:
        nos_marks += 1
        total_marks += marks
    avg_marks = total_marks / nos_marks

    if avg_marks > highest_avg_mark:
        highest_avg_mark = avg_marks
        star_stud = student['name']


print(star_stud)
print(highest_avg_mark)
        

    
```
