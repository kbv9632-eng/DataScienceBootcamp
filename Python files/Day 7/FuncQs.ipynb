```python
# Question 1 

students = [
    {
        "student_id": "101",
        "name": "Kiran",
        "age": 22,
        "gender": "Male",
        "student_class": "BSc DS"
    },
    {
        "student_id": "102",
        "name": "Alice",
        "age": 20,
        "gender": "Female",
        "student_class": "BSc AI"
    },
    {
        "student_id": "103",
        "name": "John",
        "age": 24,
        "gender": "Male",
        "student_class": "BSc DS"
    },
    {
        "student_id": "104",
        "name": "Sarah",
        "age": 19,
        "gender": "Female",
        "student_class": "BSc AI"
    }
]



def count_students_above_age():
    count = 0 
    user_inpt = input("Enter Age:- ")
    if user_inpt.strip().lower() == 'exit':
        return  
    for student in students:
        if student['age'] > user_inpt:
           count += 1
    if count > 0:
        print('There are {0} students older than {1}'.format(count,user_inpt))
    elif count == 0:
        print('No students found above the given age.')
       


count_students_above_age()
```
