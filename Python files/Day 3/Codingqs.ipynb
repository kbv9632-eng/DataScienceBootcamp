```python
# Q 14 
students = [
    {"name": "John", "marks": 85},
    {"name": "Sarah", "marks": 92},
    {"name": "Mike", "marks": 78},
    {"name": "Emma", "marks": 95}
]

total_num_dict = 0
total_marks = 0 

for student in students:
    total_num_dict += 1
    total_marks += student["marks"] 

avg_marks = total_marks / total_num_dict
print(avg_marks)
```
